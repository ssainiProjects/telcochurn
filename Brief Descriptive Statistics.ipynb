```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('c:\datasets\churn.csv')
```

```python
print("No. of rows and columns: ", df.shape)
print("\nData types:\n", df.dtypes)
```

```python
# Summary Statistics
df.describe(include='all')
```

```python
print("Value counts for Class Variable 'Churn':\n", df['Churn'].value_counts())
```

```python
# Missing Values

#'TotalCharges' is actually a numerical attriibute (it shows a total amount), but was somehow read as an object
#(possibly because of missing values).  

# Convert to numerical and check dataset for missing values

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check for any missing values in dataset after conversion
df.isnull().sum()
```

```python
#Examine correlation between the numeric features

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number])

# Calculate correlation
corr = numeric_cols.corr()

# Plot heatmap
corr_heatmap=sns.heatmap(corr, annot=True, cmap='coolwarm')
corr_heatmap
```
